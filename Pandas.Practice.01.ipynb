```python
# Change the data type of columns 'Date of Birth' from object to datetime
df['Date of Birth']=pd.to_datetime(df['Date of Birth'])
```

```python
# Recheck the data type of the columns
df.info()
```

```python
# Display the Data Set of updating the data type of Column 'Date of Birth'
df
```

```python
# Import the datetime class from datetime module
from datetime import datetime
```

```python
# Create a new columns Age by getting days from Current System Date and Date of Birth and dividing days by 365
df["Age"]=(datetime.now()-df["Date of Birth"]).dt.days//365
```

```python
# Display the current Data Frame
df
```

```python
# Create a User Defined Function da() to return Dearness Allowance (DA) based on following conditions
# 90% of basic for basic>=50000
# 80% of basic for basic>=40000
# 70% of basic for basic>=30000
# 60% for others
def da(basic):
    if   basic>=50000: return basic*0.9
    elif basic>=40000: return basic*0.8
    elif basic>=30000: return basic*0.7
    else: return basic*0.6
```

```python
# Create a new column DA by applying the user defined function da()
df['DA']=df['Basic'].apply(da)
```

```python
# Display the current Data Frame
df
```

```python
# Create a User Defined Function hra() that return House Rent Allowance (HRA) as 45% of basic
def hra(basic):
    return basic*0.45
```

```python
# Create a new column HRA by applying the user defined function hra()
df['HRA']=df['Basic'].apply(hra)
```

```python
# Display the updated Dataset
df
```

```python
# You can also create a lambda function rather than User Defined Function for small calculations
df['HRA']=df['Basic'].apply(lambda b: b*0.45)
```

```python
# Display the updated Data Frame
df
```

```python
# Create a lambda function to create a new column PF having 12% of basic
df['PF']=df['Basic'].apply(lambda b: b*0.12)
```

```python
# Display updated Data Frame
df
```

```python
# Create a new column Gross Salary by adding Basic, DA an HRA
df['Gross Salary']=df['Basic'] + df['DA'] +  df['HRA'] 
```

```python
# Display updated Data Frame
df
```

```python
# Create a new column Net Salary by  deducting PF from Gross Salary
df['Net Salary']=df['Gross Salary']-df['PF']
```

```python
# Display updated Data Frame
df
```

```python
# Display total salary, maximum salary, minimum salary and average salary paid to the employees
print('Total Salary Paid',df['Net Salary'].sum())
print('Maximum Salary Paid',df['Net Salary'].max())
print('Minimum Salary Paid',df['Net Salary'].min())
print('Average Salary Paid',df['Net Salary'].mean())
```

```python
# Display total salary paid to each department
df.groupby('Department')['Net Salary'].sum()
```

```python
# Visualize the department wise salary paid using pie chart with 6 inch by 6 inch size 
df.groupby('Department')['Net Salary'].sum().plot.pie(figsize=(6,6))
```

```python
# Visualize the department wise salary paid using bar chart with 16 inch by 4 inch size 
df.groupby('Department')['Net Salary'].sum().plot.bar(figsize=(16,4))
```

```python
# Visualize the designation wise salary for each department wise salary paid using bar chart with 16 inch by 4 inch size 
df.groupby(['Department','Designation'])['Net Salary'].sum().plot.bar(figsize=(16,4))
```
